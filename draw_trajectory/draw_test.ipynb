```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
```

```python
img_path = '../golf_data/images/750_images/750_010.jpg'

label_path = '../runs/detect/exp11/labels/750_10.txt'
```

```python
img = cv2.imread(img_path)
```

```python
height = img.shape[0]
width = img.shape[1]
print('height={}, widht={}'.format(height,width))
```

```python
plt.imshow(img)
```

```python
with open(label_path) as f:
    s = f.read()
    label_list = [float(s) for s in s.split()]
```

```python
kind, x_normal, y_normal, w_normal, h_normal, conf = label_list
```

```python
x = int(x_normal * width)
y = int(y_normal * height)
w = w_normal * width
h = h_normal * height
```

```python
#
radius = int(w / 4)
cv2.circle(img, (x, y), radius, (255, 0, 0), thickness = -1)
plt.imshow(img)
```
